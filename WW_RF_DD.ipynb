```python
import pandas as pd 

df_dd = pd.read_csv("Data/whitewine.csv")

df_dd.describe(include="all")
```

```python
from sklearn.model_selection import train_test_split

y = df_dd['high quality']
X = df_dd.drop('high quality',axis =1).drop('quality score',axis =1)

Xdd_train, Xdd_test, ydd_train, ydd_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_dd = RandomForestClassifier(n_estimators=100, random_state=42)
rfc_dd.fit(Xdd_train,ydd_train)
```

```python
rfc_dd.score(Xdd_test,ydd_test)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np 

param = {'n_estimators':[50,500],
        'max_features': range(1,11)
        }
```

```python
rf2 = RandomForestClassifier(random_state=42)
grid = GridSearchCV(estimator=rf2, param_grid=param, n_jobs=1)
grid.fit(Xdd_train,ydd_train)
```

```python
grid.best_params_
```

```python
grid.score(Xdd_test,ydd_test)
```

```python
final_model = grid.best_estimator_
```

```python
feature_imp = pd.Series(final_model.feature_importances_, index=X.columns).sort_values(ascending=False)
feature_imp
```
